```python
import gym
import minihack
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as display
```

```python
# take only chars and pixel rapresentations of the map

env = gym.make("MiniHack-ExploreMaze-Easy-Mapped-v0", observation_keys=("chars", "pixel"))
state = env.reset()
env.render()
```

```python
# state['chars'] let you see the map level as multi-dim array of ascii codes 
# "@" is the position of the player, and his ascii code is 64
# ">" is the position of the stairs, and his ascii code is 62
 
# with this you can find in the map (visualized as "chars") the stairs location
y0, x0 = np.where(state['chars'] == ord(">"))
# with this you can find in the map the player location
y1, x1 = np.where(state['chars'] == ord("@"))

print(f' le coordinate delle scale sono: y:{y0} la x:{x0}')
print(f' le coordinate del player sono: y:{y1} la x:{x1}')
```
